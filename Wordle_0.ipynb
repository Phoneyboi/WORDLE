```python
#CREATES A LIST OF LETTERS THAT SPELL OUT THE 5-LETTER WORD
word_list = [] 

# STRING OF LETTERS IN ALPHABET TO CHOOSE FROM IN CREATING A WORD
letters = 'abcdefghijklmnopqrstuvwxyz'

# CREATES A LIST OF WORDS THAT HAVE BEEN GENERATED
list_of_words = []

# VARIABLE TO ADJUST HOW MANY WORD OPTIONS USER GETS AFTER RUNNING A COMMAND
word_options = 10

# INPUT A LETTER THAT IS KNOWN TO BE IN THE WORD, BUT LOCATION IS UNKNOWN
ness_let_1 = 's'
ness_loc_1 = 0

#ness_let_2 = 'i'
#ness_loc_2 = 1

#ness_let_3 = 'u'
#ness_loc_3 = 0

# THIS IS A SECTION FOR LIST OF LETTERS THAT ARE NOT IN THE WORD. STILL DEVELOPING.
#bad_let =

# INPUT THE LETTERS WHOSE LOCATION IN THE STRING IN KNOWN
a = 's'
b = 't'
c = ' '
d = ' '
e = ' '

# ITERABLE VARIABLE
i = 0 

# WHILE FUNCTION IS LENGTH OF WORD (5) PLUS ONE TO RUN A CHECK COMMAND AT THE END
while i < 5:
    
    if i == 0: # For each letter in the word
        if a == ' ': # Check if string has been found
            word_list.append(random.choice(letters))
        else:
            word_list.append(a)
        i = i + 1
        
    if i == 1:
        if b == ' ':
            word_list.append(random.choice(letters))
        else:
            word_list.append(b)
        i = i + 1
        
    if i == 2:
        if c == ' ':
            word_list.append(random.choice(letters))
        else:
            word_list.append(c)
        i = i + 1
        
    if i == 3:
        if d == ' ':
            word_list.append(random.choice(letters))
        else:
            word_list.append(d)
        i = i + 1
        
    if i == 4:
        if e == ' ':
            word_list.append(random.choice(letters))
        else:
            word_list.append(e)
        i = i + 1
        
    if i == 5:    
        # CHECK IF LETTERS ARE IN THE WORD_LIST
        if ness_let_1 not in ''.join(word_list) : #or word_list[ness_loc_1] == ness_let_1
            word_list = []
            i = 0
        #if ness_let_2 not in ''.join(word_list) or word_list[ness_loc_2] == ness_let_2:
            #word_list = []
            #i = 0   
        #if ness_let_3 not in ''.join(word_list) or word_list[ness_loc_3] == ness_let_3:
            #word_list = []
            #i = 0  
        #if bad_let in ''.join(word_list):
            #word_list = []
            #i = 0
        else:
            word = ''.join(word_list) 
            if word in english_words_lower_alpha_set: # Check if word is a word
                list_of_words.append(word)
            if list_of_words.count(word) == 2: # Check to see if word has already been recognized
                # Instead of removing every instance of the word in the list, the below function 
                # will only remove the last word appended in the list
                list_of_words.remove(list_of_words[-1]) 
                word_list = []
                i = 0
            if len(list_of_words) == word_options:
                break
            else:
                word_list = []
                i = 0 

# PRINTS THE LIST OF WORDS GENERATED BY THE LOOP
print(list_of_words)
```
